### Load Edgar files

```python
df_edgar1 = pd.read_feather(data_folder + 'stock_edgar_1_of_3.feather')
df_edgar2 = pd.read_feather(data_folder + 'stock_edgar_2_of_3.feather')
df_edgar3 = pd.read_feather(data_folder + 'stock_edgar_3_of_3.feather')
```

```python
df_edgar1.sample(5)
```

```python
print(sorted(df_edgar1[df_edgar1.type=='Year'].frame.unique()))
print(sorted(df_edgar2[df_edgar2.type=='Year'].frame.unique()))
print(sorted(df_edgar3[df_edgar3.type=='Year'].frame.unique()))
```

```python
# Combine the dataframes to get the data for the the full year 2023
df_edgar_CY2023 = pd.concat([df_edgar1[df_edgar1.frame=='CY2023'],
          df_edgar2[df_edgar2.frame=='CY2023'],
          df_edgar3[df_edgar3.frame=='CY2023']], axis=0)

# Transpose measure to columns
df_edgar_CY2023_pivot = df_edgar_CY2023.pivot(index=['sym', 'type', 'frame'], columns='measure', values='value').reset_index()
df_edgar_CY2023_pivot.columns.name = None
df_edgar_CY2023_pivot.head()
```

```python
# Find fill rate for each column in CY2023 data
fill_rate = pd.DataFrame(df_edgar_CY2023_pivot.count()/len(df_edgar_CY2023_pivot), columns=['fill_rate']).reset_index().rename(columns={'index':'column'}).sort_values(by='fill_rate', ascending=False)
print(f"Number of measures in CY 2023 files = {len(fill_rate)-3}")
print("Features with >80% fill rate are as follows:")
display(fill_rate[(fill_rate.fill_rate>0.8)])
# Keep only the columns with >80% fill rate
columns_to_keep_CY2023 = fill_rate[(fill_rate.fill_rate>0.8)].column.tolist()
```

```python
# Combine the dataframes to get the data for the Q1 of 2024 for both the quarterly and Instantaneous reports
df_edgar_2024Q1 = pd.concat([df_edgar1[((df_edgar1.frame=='CY2024Q1I')| (df_edgar1.frame=='CY2024Q1'))],
          df_edgar2[((df_edgar2.frame=='CY2024Q1I')| (df_edgar2.frame=='CY2024Q1'))],
          df_edgar3[((df_edgar3.frame=='CY2024Q1I')| (df_edgar3.frame=='CY2024Q1'))]], axis=0)
```

```python
# Transpose measure to columns
df_edgar_2024Q1_pivot = df_edgar_2024Q1.pivot(index=['sym', 'type', 'frame'], columns='measure', values='value').reset_index()
df_edgar_2024Q1_pivot.columns.name = None

# Removing the instantaneous reports to keep only the full quarterly reports
df_edgar_2024Q1_pivot_fullqtr=df_edgar_2024Q1_pivot[df_edgar_2024Q1_pivot.frame=='CY2024Q1']
# Find fill rate for each column
fill_rate = pd.DataFrame(df_edgar_2024Q1_pivot_fullqtr.count()/len(df_edgar_2024Q1_pivot_fullqtr), columns=['fill_rate']).reset_index().rename(columns={'index':'column'}).sort_values(by='fill_rate', ascending=False)
print(f"Number of measures in 2024 Q1 Full quarter file = {len(fill_rate)-3} with {len(df_edgar_2024Q1_pivot_fullqtr)} records")
print("Features with >80% fill rate are as follows:")
display(fill_rate[(fill_rate.fill_rate>0.8)])
# Keep only the columns with >80% fill rate
columns_to_keep_2024Q1 = fill_rate[(fill_rate.fill_rate>0.8)].column.tolist()

# Keeping only the instantaneous reports
df_edgar_2024Q1_pivot_instantaneous=df_edgar_2024Q1_pivot[df_edgar_2024Q1_pivot.frame=='CY2024Q1I']
# Find fill rate for each column
fill_rate = pd.DataFrame(df_edgar_2024Q1_pivot_instantaneous.count()/len(df_edgar_2024Q1_pivot_instantaneous), columns=['fill_rate']).reset_index().rename(columns={'index':'column'}).sort_values(by='fill_rate', ascending=False)
print(f"Number of measures in 2024 Q1 instantaneous quarter file = {len(fill_rate)-3} with {len(df_edgar_2024Q1_pivot_instantaneous)} records")
print("Features with >80% fill rate are as follows:")
fill_rate[(fill_rate.fill_rate>0.8)]
# Keep only the columns with >80% fill rate
columns_to_keep_2024Q1I = fill_rate[(fill_rate.fill_rate>0.8)].column.tolist()
```

##### <font color="FFA07A"> Finding: Yearly, Quarterly and Instantaneous frames have different metrics. It's NOT the same metrics for different time frames!!!</font>

```python
# Keeping columns with >80% fill rate
df_edgar_CY2023_pivot_filled_80 = df_edgar_CY2023_pivot[columns_to_keep_CY2023]
df_edgar_2024Q1_pivot_fullqtr_filled_80 = df_edgar_2024Q1_pivot_fullqtr[
    columns_to_keep_2024Q1
]
df_edgar_2024Q1_pivot_instantaneous_filled_80 = df_edgar_2024Q1_pivot_instantaneous[
    columns_to_keep_2024Q1I
]

# Combine the dataframes to get merged data for CY2023, 2024Q1 and 2024Q1I data
df_edgar_combined = pd.merge(
        df_edgar_CY2023_pivot_filled_80.drop(columns=["type", "frame"]),
        df_edgar_2024Q1_pivot_fullqtr_filled_80.drop(columns=["type", "frame"]),
        on="sym",
        how="outer",
        suffixes=("_CY2023", "_2024Q1"),
    ).merge(df_edgar_2024Q1_pivot_instantaneous_filled_80.drop(columns=["type", "frame"]), on="sym", how="outer", suffixes=("", "_2024Q1I")
)

df_edgar_combined.head()
```

```python
# Save the 2023, 2024Q1 and 2024Q1I merged file as a pickle files
df_edgar_combined.to_pickle(data_folder + 'df_edgar_combined.pkl')
```

#### Testing if historical data can be used to impute values

```python
# Missing value imputation for the CY2023 data
df_edgar_CY2023_pivot_filled_80[df_edgar_CY2023_pivot_filled_80.isnull().any(axis=1)].head(5)
```

```python
# All the yearly data for relevant columns merged
df_edgar_yearly=pd.concat([
    df_edgar1[(df_edgar1.type=='Year') & (df_edgar1.measure.isin(list(df_edgar_CY2023_pivot_filled_80.columns)))],
    df_edgar2[(df_edgar2.type=='Year') & (df_edgar2.measure.isin(list(df_edgar_CY2023_pivot_filled_80.columns)))],
    df_edgar3[(df_edgar3.type=='Year') & (df_edgar3.measure.isin(list(df_edgar_CY2023_pivot_filled_80.columns)))]
], axis=0).reset_index(drop=True)
df_edgar_yearly.tail()
```

```python
# Test if we can impute missing values for the CY2023 data
for column in df_edgar_CY2023_pivot_filled_80.columns:
    for index, row in df_edgar_CY2023_pivot_filled_80[df_edgar_CY2023_pivot_filled_80[column].isnull()].iterrows():
        if (df_edgar_yearly[(df_edgar_yearly.sym=='CPT')&(df_edgar_yearly.measure=='IncomeTaxExpenseBenefit') & (df_edgar_yearly.type=='Year')]).shape[0]>0:
            print(f"Missing value in {column} for {row['sym']} can be imputed with {df_edgar_yearly[(df_edgar_yearly.sym==row['sym'])&(df_edgar_yearly.measure==column) & (df_edgar_yearly.type=='Year')].iloc[0]['value']}")
```

#### Extracting sector data

```python
# Load the data from req URL with urllib
#ssl._create_default_https_context = ssl._create_unverified_context
req='https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
context = ssl._create_unverified_context()
html = urllib.request.urlopen(req,context=context).read()

# Read the table from the URL
df_stock = pd.read_html(html)[0]

# Make sure CIK is a string with 10 characters (leading zeros)
df_stock['CIK'] = df_stock['CIK'].astype(str).str.zfill(10)

# Save the data to a pickle file
df_stock.to_pickle(data_folder + 'df_stock.pkl')

df_stock.tail()
```
